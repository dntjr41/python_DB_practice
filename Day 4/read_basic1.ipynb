```python
import pandas as pd
```

```python
mysource = {'시도':['서울', '경기', '인천', '부산', '대전'],
           '구분':['특별시', '도', '광역시', '광역시', '광역시'],
           '인구':['990만명', '1300만명', '350만명', '300만명', '150만명'],
           '면적':[605.2, "10,171", "1,029", 767.4, 539.8]}
```

```python
df = pd.DataFrame(mysource)
df
```

```python
# csv파일로 저장하는 방법
df.to_csv('sample1031.csv')
```

```python
# csv파일을 로딩하는 방법
df1 = pd.read_csv('sample1031.csv')
df1
```

```python
#컬럼명을 지정
df1 = pd.read_csv('sample1031.csv', names = ['no', '시도', ' 구분', '인구', '면적'])
df1
```

```python
#컬럼명을 인덱스로 지정
df1 = pd.read_csv('sample1031.csv', index_col = 'no', names = ['no', '시도', ' 구분', '인구', '면적'])
df1
```

```python
#불필요한 행 제외하고 로딩시키기
df1 = pd.read_csv('sample1031.csv', index_col = 'no',
                  names = ['no', '시도', ' 구분', '인구', '면적'],
                 skiprows = [0])
df1
```
