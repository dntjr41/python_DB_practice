```python
data = pd.read_csv('accidentdata.csv')
data.head(5)
```

```python
d2 = data[(data.사상자수>=3)]
d2
```

```python
d3 = d2.groupby('요일')
d3
```

```python
result = d3['사상자수'].sum().plot(kind='bar', title='2012-2014 요일별 중대교통사고 사상자 분석')
result.set_xticklabels(['금요일', '목요일', '수요일', '월요일', '일요일', '토요일', '화요일'], fontsize=10, rotation=0)
```
