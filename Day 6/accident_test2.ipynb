```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
```

```python
#한글깨짐 방지작업
font_info="C:/Users/LG/Jupyter workspace/Day 6/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_info).get_name()
rc('font', family=font_name)
```

```python
data = pd.read_csv('accidentdata.csv')
```

```python
df1 = data[data.발생지시도=='경기']
df1
```

```python
df2 = df1.groupby(['발생지시군구'])
df2
```

```python
df3 = df2['사망자수'].sum()
df3
```

```python
result = df3.sort_values(ascending=True).tail(5)
result
```

```python
result.plot(kind='pie',title='2012-2014 경기도내 교통사망사고 빈번지역')
```

```python
#이렇게도 작성 가능. matplotlib.pyplot 이용하여 파이차트 작성
c = ['red','brown','green','purple','yellow']
plt.pie(result, colors=c, autopct='%.2f')
plt.title('2012-2014 경기도내 교통사망사고 빈발지역')
```
