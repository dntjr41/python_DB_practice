```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy.random as nrp
```

```python
plt.plot(nrp.randn(10))
plt.xlabel('data_x')
plt.ylabel('value_y')
plt.title('just sample')
plt.text(0.5, 0.5, 'random value')
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
a1 = fig.add_subplot(1,1,1)
a1.plot(nrp.randn(40), 'r-+')
a1.set_xticks([10,20,30,40,50])
a1.set_xticklabels(['서울','경기','인천','강원','세종'])
```
